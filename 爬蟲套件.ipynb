## 由Google News擷取新聞 - scraparazzie

```python
!pip install scraparazzie #安裝套件
from scraparazzie import scraparazzie
```

```python
client = scraparazzie.NewsClient(language='chinese traditional', location='Taiwan', topic='Business', max_results=20) #爬取台灣的中文商業新聞
client.print_news()
```
